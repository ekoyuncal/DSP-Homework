### 2-)
Bir önceki soruyu 𝐹𝑠 = 1000 𝐻𝑧 için tekrarlayınız.
FN = 100 Fs = 1000  N -> 10k N = 10



```python
#Örnekleme frekansı ve periyodunun tanımlanması
Fs = 1000
Ts = 1/Fs

N = 10

#sinyallerin ve n örnek sayısının tanımlanması
n = np.arange(0, 2*N)
x1n = 7*np.cos(200*np.pi*n*Ts)
x2n = 3*np.cos(200*np.pi*n*Ts) + 4*np.cos(1800*np.pi*n*Ts)

#Grafiklerin çizdirilmesi
plt.figure()
plt.stem(n,x1n)
plt.title("x1[n]") 
plt.xlabel("örnek") 
plt.ylabel("x[n]") 

plt.figure()
plt.stem(n,x2n)
plt.title("x2[n]") 
plt.xlabel("örnek") 
plt.ylabel("x[n]") 
```

```python
#omeganın -pi ve pi aralıgında tanımlanması
w = np.linspace(-np.pi,np.pi-(2*np.pi/N), N)

#işaretin fourier transformu ve -pi ile pi aralığında gösterilmesi
x1_w = np.fft.fft(x1n,N)/N 
x1_w = np.fft.fftshift(x1_w)

x2_w = np.fft.fft(x2n,N)/N
x2_w = np.fft.fftshift(x2_w)

#Grafiklerin çizdirilmesi
plt.figure()
plt.stem(w, np.abs(x1_w))
plt.title("X1(w) Amplitude Grafigi")
plt.xlabel("w")

plt.figure()
plt.stem(w,np.angle(x1_w))
plt.title("X1(w) Faz Grafigi")
plt.xlabel("w")

plt.figure()
plt.stem(w, np.abs(x2_w))
plt.title("X2(w) Amplitude Grafigi")
plt.xlabel("w)")

plt.figure()
plt.stem(w,np.angle(x2_w))
plt.title("X2(w) Faz Grafigi")
plt.xlabel("w")
```

𝑥3(𝑡) = 𝑠𝑖𝑛(9600𝜋𝑡) işaretini 𝐹𝑠 = 8192 𝐻𝑧 ile örnekleyerek 𝑥3[𝑛] işaretini elde edilmektedir.
a. 𝑥3(𝑡) işaretini tam 3 periyot için çizdiriniz.


b. 𝑥3[𝑛] işaretini tam 3 periyot için çizdiriniz.


c. 𝑥3[𝑛] işaretinin fft() fonksiyonunu kullanarak Fourier dönüşümünü elde ediniz ve genlik ve faz grafiklerini çizdiriniz.


d. (c) şıkkında elde ettiğiniz grafikten yararlanarak bu grafiğin hangi fonksiyona ait olduğunu bulunuz (Not: 𝑥4[𝑛] = 𝐴𝑠𝑖𝑛(𝐵𝜋𝑛) formundaki işaret için 𝐴 ve 𝐵 parametrelerini bulacaksınız).


e. 𝑥3[𝑛] ile 𝑥4[𝑛]’i aynı grafik üzerinde çizdirip sonucu yorumlayınız. 



a-) işlemi N hesaplaması
FN = 4800 FS = 8192  N-> 8192/4800 = 128/75 k k=75 N=128

```python
#Örnekleme frekansı ve periyodunun tanımlanması
Fn = 4800
Fs = 8192

Tn = 1/Fn
Ts = 1/Fs

N = 128

#3 periyot olarak t ve sinyalin tanımlanması
t = np.arange(0, 3*Tn, 0.000001)
x3_t = np.sin(9600*np.pi*t)

# Sürekli Grafiğin çizdirilmesi
plt.figure()
plt.plot(t, x3_t)
plt.xlabel("t")
plt.title("x3(t)");

#ayrık zamanlı örneklenmiş sinyalin tanımlanması
n = np.arange(0, 3*N)
x3_n = np.sin(9600*np.pi*n*Ts)

#Örneklenmiş Grafiği çizdirilmesi
plt.figure()
plt.stem(n, x3_n)
plt.xlabel("n")
plt.title("x3[n]");

#x3[n] işaretinin fft ile elde edilmesi ve genlik ile faz grafiklerinin çizdirilmesi

#omeganın -pi ve pi aralıgında tanımlanması
w = np.linspace(-np.pi,np.pi-(2*np.pi/N), N)

#işaretin fourier transformu ve -pi ile pi aralığında gösterilmesi
x3_w = np.fft.fft(x3_n,N)/N 
x3_w = np.fft.fftshift(x3_w)

#Grafiklerin çizdirilmesi (Sırasıyla ; Genlik ve Faz Grafikleri)
plt.figure()
plt.stem(w, np.abs(x3_w))
plt.title("X3(w) Amplitude Grafigi")
plt.xlabel("w")

plt.figure()
plt.stem(w,np.angle(x3_w))
plt.title("X3(w) Faz Grafigi")
plt.xlabel("w")
```

### 4-)
𝑥[𝑛] =[1.25 , 3.5 , 4 , 6.7 , −1.4 , −3.5 , −10.5 , 8.5 , 3.8] olarak verilmektedir.𝑥[𝑛] işareti 3bit ile ifade edilmek isteniyorsa, mid-rise ve mid-tread yöntemleri ile işareti kuantalayarak çizdiriniz.

```python
#işaret ve örnek sayısının tanımlanması
xn = np.array([1.25 , 3.5 , 4 , 6.7 , -1.4 , -3.5 , -10.5 , 8.5 , 3.8])
n = np.arange(0, 9)

#Kuantalama fonksiyonunun tanımlanması
def quantize(x, S):
    X = x.reshape((-1,1))
    S = S.reshape((1,-1))
    dists = abs(X-S)
 
    nearestIndex = dists.argmin(axis=1)
    quantized = S.flat[nearestIndex]
 
    return quantized.reshape(x.shape)

# işaret değerlerinin kaç bit ile ifade edileceği
k_bit= 3
# işaretin y ekseninde aldığı max değer aralığı
Max_range= np.amax(xn)-np.amin(xn)
# seviye sayısı
q=pow(2,k_bit) 
# adım aralığı
delta = Max_range/q

# mid-rise yöntemi için formül
S_midrise = -np.amax(xn) + delta/2 + np.arange(q)*delta 
# mid-tread yöntemi için formül
S_midtread = -np.amax(xn) + np.arange(q)*delta 

# mid-rise yöntemi ile kuantalanmış değerler
y_midtrise = quantize(xn, S_midrise) 
# mid-tread yöntemi ile kuantalanmış değerler
y_midtread = quantize(xn, S_midtread)

#Grafiklerin yöntemlere göre çizdirilmesi
plt.figure()
plt.stem(n, xn)
plt.bar(n,y_midtrise,color='orange',edgecolor = "darkorange")
plt.xlabel("örnek") 
plt.title('midrise yöntemiyle kuantalanmış işaret')

plt.figure()
plt.stem(n, xn)
plt.bar(n,y_midtread,color='orange',edgecolor = "darkorange")
plt.xlabel("örnek")
plt.title('midtread yöntemiyle kuantalanmış işaret')
```

```python
#Ã–rnekleme frekansÄ± ve periyodunun tanÄ±mlanmasÄ±
Fs = 1000
Ts = 1/Fs

N = 10

#sinyallerin ve n Ã¶rnek sayÄ±sÄ±nÄ±n tanÄ±mlanmasÄ±
n = np.arange(0, 2*N)
x1n = 7*np.cos(200*np.pi*n*Ts)
x2n = 3*np.cos(200*np.pi*n*Ts) + 4*np.cos(1800*np.pi*n*Ts)

#Grafiklerin Ã§izdirilmesi
plt.figure()
plt.stem(n,x1n)
plt.title("x1[n]") 
plt.xlabel("Ã¶rnek") 
plt.ylabel("x[n]") 

plt.figure()
plt.stem(n,x2n)
plt.title("x2[n]") 
plt.xlabel("Ã¶rnek") 
plt.ylabel("x[n]") 
```

```python
#omeganÄ±n -pi ve pi aralÄ±gÄ±nda tanÄ±mlanmasÄ±
w = np.linspace(-np.pi,np.pi-(2*np.pi/N), N)

#iÅŸaretin fourier transformu ve -pi ile pi aralÄ±ÄŸÄ±nda gÃ¶sterilmesi
x1_w = np.fft.fft(x1n,N)/N 
x1_w = np.fft.fftshift(x1_w)

x2_w = np.fft.fft(x2n,N)/N
x2_w = np.fft.fftshift(x2_w)

#Grafiklerin Ã§izdirilmesi
plt.figure()
plt.stem(w, np.abs(x1_w))
plt.title("X1(w) Amplitude Grafigi")
plt.xlabel("w")

plt.figure()
plt.stem(w,np.angle(x1_w))
plt.title("X1(w) Faz Grafigi")
plt.xlabel("w")

plt.figure()
plt.stem(w, np.abs(x2_w))
plt.title("X2(w) Amplitude Grafigi")
plt.xlabel("w)")

plt.figure()
plt.stem(w,np.angle(x2_w))
plt.title("X2(w) Faz Grafigi")
plt.xlabel("w")
```

```python
#Ã–rnekleme frekansÄ± ve periyodunun tanÄ±mlanmasÄ±
Fn = 4800
Fs = 8192

Tn = 1/Fn
Ts = 1/Fs

N = 128

#3 periyot olarak t ve sinyalin tanÄ±mlanmasÄ±
t = np.arange(0, 3*Tn, 0.000001)
x3_t = np.sin(9600*np.pi*t)

# SÃ¼rekli GrafiÄŸin Ã§izdirilmesi
plt.figure()
plt.plot(t, x3_t)
plt.xlabel("t")
plt.title("x3(t)");

#ayrÄ±k zamanlÄ± Ã¶rneklenmiÅŸ sinyalin tanÄ±mlanmasÄ±
n = np.arange(0, 3*N)
x3_n = np.sin(9600*np.pi*n*Ts)

#Ã–rneklenmiÅŸ GrafiÄŸi Ã§izdirilmesi
plt.figure()
plt.stem(n, x3_n)
plt.xlabel("n")
plt.title("x3[n]");

#x3[n] iÅŸaretinin fft ile elde edilmesi ve genlik ile faz grafiklerinin Ã§izdirilmesi

#omeganÄ±n -pi ve pi aralÄ±gÄ±nda tanÄ±mlanmasÄ±
w = np.linspace(-np.pi,np.pi-(2*np.pi/N), N)

#iÅŸaretin fourier transformu ve -pi ile pi aralÄ±ÄŸÄ±nda gÃ¶sterilmesi
x3_w = np.fft.fft(x3_n,N)/N 
x3_w = np.fft.fftshift(x3_w)

#Grafiklerin Ã§izdirilmesi (SÄ±rasÄ±yla ; Genlik ve Faz Grafikleri)
plt.figure()
plt.stem(w, np.abs(x3_w))
plt.title("X3(w) Amplitude Grafigi")
plt.xlabel("w")

plt.figure()
plt.stem(w,np.angle(x3_w))
plt.title("X3(w) Faz Grafigi")
plt.xlabel("w")
```

### 4-)
ð‘¥[ð‘›] =[1.25 , 3.5 , 4 , 6.7 , âˆ’1.4 , âˆ’3.5 , âˆ’10.5 , 8.5 , 3.8] olarak verilmektedir.ð‘¥[ð‘›] iÅŸareti 3bit ile ifade edilmek isteniyorsa, mid-rise ve mid-tread yÃ¶ntemleri ile iÅŸareti kuantalayarak Ã§izdiriniz.

```python
#iÅŸaret ve Ã¶rnek sayÄ±sÄ±nÄ±n tanÄ±mlanmasÄ±
xn = np.array([1.25 , 3.5 , 4 , 6.7 , -1.4 , -3.5 , -10.5 , 8.5 , 3.8])
n = np.arange(0, 9)

#Kuantalama fonksiyonunun tanÄ±mlanmasÄ±
def quantize(x, S):
    X = x.reshape((-1,1))
    S = S.reshape((1,-1))
    dists = abs(X-S)
 
    nearestIndex = dists.argmin(axis=1)
    quantized = S.flat[nearestIndex]
 
    return quantized.reshape(x.shape)

# iÅŸaret deÄŸerlerinin kaÃ§ bit ile ifade edileceÄŸi
k_bit= 3
# iÅŸaretin y ekseninde aldÄ±ÄŸÄ± max deÄŸer aralÄ±ÄŸÄ±
Max_range= np.amax(xn)-np.amin(xn)
# seviye sayÄ±sÄ±
q=pow(2,k_bit) 
# adÄ±m aralÄ±ÄŸÄ±
delta = Max_range/q

# mid-rise yÃ¶ntemi iÃ§in formÃ¼l
S_midrise = -np.amax(xn) + delta/2 + np.arange(q)*delta 
# mid-tread yÃ¶ntemi iÃ§in formÃ¼l
S_midtread = -np.amax(xn) + np.arange(q)*delta 

# mid-rise yÃ¶ntemi ile kuantalanmÄ±ÅŸ deÄŸerler
y_midtrise = quantize(xn, S_midrise) 
# mid-tread yÃ¶ntemi ile kuantalanmÄ±ÅŸ deÄŸerler
y_midtread = quantize(xn, S_midtread)

#Grafiklerin yÃ¶ntemlere gÃ¶re Ã§izdirilmesi
plt.figure()
plt.stem(n, xn)
plt.bar(n,y_midtrise,color='orange',edgecolor = "darkorange")
plt.xlabel("Ã¶rnek") 
plt.title('midrise yÃ¶ntemiyle kuantalanmÄ±ÅŸ iÅŸaret')

plt.figure()
plt.stem(n, xn)
plt.bar(n,y_midtread,color='orange',edgecolor = "darkorange")
plt.xlabel("Ã¶rnek")
plt.title('midtread yÃ¶ntemiyle kuantalanmÄ±ÅŸ iÅŸaret')
```

```python
#2 Periyot olacak şekilde isteniyor
#2pi/2pi/5 ilk sinyalden gelen5
#2pi/3pi/5 10k/3 k=3 ikinci sinyalden gelen 10 
#Bu sinyallerin ekoku sonucu periyot 10
#2 tam periyot için
T = 10
n = np.arange(0,2*T)

signalx =  2*np.sin(2*np.pi/5*n - np.pi/5) + np.sin(3*np.pi/5*n - 2*np.pi/5)

plt.stem(n, signalx)
plt.title("Discrete Time Signal")

#Periyot testi
plt.axvspan(0 , T-1, color="#EED1CE", alpha=0.3)
plt.axvspan(T, 2*T-1, color="#535151", alpha=0.3)
plt.text(7.5, 1.5, "Period N = 10", fontsize = 12,
 bbox = dict(facecolor = "#F6C9C9", alpha = 0.5))
```

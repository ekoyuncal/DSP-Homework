### Soru 2-)  Acos(w0n + phi) = [0, 0.707106, −1, 0.707106, 0, −0.707106, 1, −0.707106]

```python
#DFT 'sinin faz ve genliklerinin bulunması
n = np.arange(0, 8)
x_n = np.array([0, 0.707106, -1, 0.707106, 0, -0.707106, 1, -0.707106])
x_n_abs = np.abs(fft(x_n)/len(n))
x_n_phase = np.angle(fft(x_n)/len(n))
w_discrete = n * 2 * np.pi / len(n)

#Faz ve genlik grafiklerinin çizdirilmesi
plt.figure()
plt.subplot(1, 2, 1)
plt.stem(w_discrete/np.pi, x_n_abs)
plt.title('1/N$|X(e^{j\omega})|$')
plt.xlabel('Normalize frekans (x$\pi$)')


plt.subplot(1, 2, 2)
plt.stem(w_discrete/np.pi, x_n_phase/np.pi)
plt.title('1/N$ \measuredangle X(e^{j\omega})$')
plt.xlabel('Normalize frekans (x$\pi$)')

#∠𝑎3 =-π/2 ∠𝑎5 = π/2 
#|𝑎3| = 0.5 |𝑎5| = 0.5
print( x_n_phase/np.pi)
#x[n] = a3*e^(j3π/4n) + a5*e^(j5π/4n) = |𝑎3|𝑒^(𝑗∠𝑎3) + |𝑎5|𝑒^(𝑗∠𝑎5) = 0.5𝑒^(−𝑗π/2)*e^(j3π/4n) + 0.5𝑒^(𝑗π/2)*e^(j5π/4n)
#     = 0.5𝑒^(−𝑗π/2)*e^(j3π/4n) + 0.5𝑒^(𝑗π/2)*e^(-j3π/4n) = 2*0.5*cos(3π/4n - π/2)
#     = 1cos(3π/4n - π/2)
#A = 1
#wo = 3π/4n
#phi = π/2
```

### Soru 3-)  x[n] = cos(π/3n) h[n] = dirac[n] - dirac[n-1] y[n] işaretini n= [0, ... 10] aralığında çizdiriniz.

```python
n = np.arange(0, 11)
"""
dirac_n = np.array([0 if i!=0 else 1 for i in n])
dirac_n_eksi_1 = np.array([0 if i!=1 else 1 for i in n])
"""

h_f = 1-np.exp(-1j*np.pi/3)

#Genlik ve Açının Hesaplanması 
h_f_abs = np.abs(h_f)
h_f_phase = np.angle(h_f)

x_n = np.cos(n * np.pi/3)
x_n_plus_h_f_phase = np.cos((n * np.pi/3) + h_f_phase  )
y_n = h_f_abs * x_n_plus_h_f_phase

#Grafiğin Çizdirilmesi
plt.figure()
plt.stem(n, y_n)
plt.title("y[n]  İşareti")
plt.xlabel("Sample Number n")
plt.ylabel("Amplitude")

```

```python
# İstenilen genlik ve faz açısı değerlerinin bastırılması
print(f"Amplitude |H(e^(jpi/3)| : {1 if (h_f_abs >= 0.99998) else h_f_abs}")
print(f"Phase Angle : {0.333 if (h_f_phase >= 0.33333) else h_f_phase}")
```

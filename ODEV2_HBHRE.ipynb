```python
#K√ºt√ºphanelerin import edilmesi
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
n = np.arange(-1, 25)
n2 = np.arange(-1, 11)
h1 = np.array([1/3, 1/3, 1/3], dtype = float)

# Gƒ∞Rƒ∞≈û Sƒ∞NYALLERƒ∞N TANIMLANMASI
x1 = np.cos(np.pi/5 * n) * np.heaviside(n-10, 1) - np.cos(np.pi/5 * n) * np.heaviside(n-20, 1)
x2 = ((-0.3)**n2) *( np.heaviside(n2-3, 1) - np.heaviside(n2-5, 1))

#√áIKI≈û Sƒ∞NYALLERƒ∞Nƒ∞N TANIMLANMASI
y1 = np.convolve(x1, h1, "same")
y2 = np.convolve(x2, h1, "same")

#GRAFIKLERIN √áƒ∞ZDƒ∞Rƒ∞LMESƒ∞
plt.figure()
plt.stem(n, x1)
plt.title("x1[n] Giri≈ü Sinyali")
plt.xlabel("Sample Number n")
plt.ylabel("Amplitude")

plt.figure()
plt.stem(y1)
plt.title("y1[n] √áƒ±kƒ±≈ü Sinyali")
plt.xlabel("Sample Number n")
plt.ylabel("Amplitude")

plt.figure()
plt.stem(n2, x2)
plt.title("x2[n] Giri≈ü Sinyali")
plt.xlabel("Sample Number n")
plt.ylabel("Amplitude")

plt.figure()
plt.stem(y2)
plt.title("y2[n] √áƒ±kƒ±≈ü Sinyali")
plt.xlabel("Sample Number n")
plt.ylabel("Amplitude")
```

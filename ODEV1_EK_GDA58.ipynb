```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook
```

### Soru 2

```python
T = 10 #t1 ve t2 10

n = np.arange(0,2*T)

signalx1 =  2*np.sin(2*np.pi/5*n - np.pi/5)
signalx2 =  np.sin(3*np.pi/5*n - 2*np.pi/5)



plt.stem(n, signalx1)
plt.title("Discrete Time Signal I")

#Periyot testi
plt.axvspan(0 , T-1, color="#EED1CE", alpha=0.3)
plt.axvspan(T , 2*T-1, color="#535151", alpha=0.3)
plt.text(7.5, 1.5, "Period N = 10", fontsize = 12,
 bbox = dict(facecolor = "#F6C9C9", alpha = 0.5))

plt.figure()
plt.stem(n, signalx2)
plt.title("Discrete Time Signal II")

#Periyot testi
plt.axvspan(0 , T-1, color="#EED1CE", alpha=0.3)
plt.axvspan(T , 2*T-1, color="#535151", alpha=0.3)
plt.text(7.5, 0.75, "Period N = 10", fontsize = 12,
 bbox = dict(facecolor = "#F6C9C9", alpha = 0.5))
```

### Soru 3

```python

T = 10 #T tÃ¼m sinyaller iÃ§in 10

n1 = np.arange(0, 16)
n2 = np.arange(-10, 5)


signal1 = np.exp(np.pi/5)
signal3 = np.exp(1j*n1*np.pi/5)
signal4 = np.cos(np.pi/5) + 1j*np.sin(np.pi/5)
signal5 = np.cos(n1*np.pi/5) + 1j*np.sin(n1*np.pi/5)
signal6 = np.cos(n2*np.pi/5) + 1j*np.sin(n2*np.pi/5)


fig = plt.figure()
reel=fig.add_subplot(211)
imag=fig.add_subplot(212)
reel.stem(n1, signal1.real*np.ones(16))
reel.title.set_text(" x1[n] Reel")
imag.stem(n1, signal1.imag*np.ones(16))
imag.title.set_text(" x1[n] Imag")
plt.tight_layout(pad=2.0)


fig2 = plt.figure()
reel=fig2.add_subplot(211)
imag=fig2.add_subplot(212)
reel.stem(n1, signal3.real)
reel.title.set_text(" x3[n] Reel")
imag.stem(n1, signal3.imag)
imag.title.set_text(" x3[n] Imag")
plt.tight_layout(pad=2.0)

fig3 = plt.figure()
reel=fig3.add_subplot(211)
imag=fig3.add_subplot(212)
reel.stem(n1, signal4.real*np.ones(16))
reel.title.set_text(" x4[n] Reel")
imag.stem(n1, signal4.imag*np.ones(16))
imag.title.set_text(" x4[n] Imag")
plt.tight_layout(pad=2.0)

fig4 = plt.figure()
reel=fig4.add_subplot(211)
imag=fig4.add_subplot(212)
reel.stem(n1, signal5.real)
reel.title.set_text(" x5[n] Reel")
imag.stem(n1, signal5.imag)
imag.title.set_text(" x5[n] Imag")
plt.tight_layout(pad=2.0)


fig5 = plt.figure()
reel=fig5.add_subplot(211)
imag=fig5.add_subplot(212)
reel.stem(n2, signal6.real)
reel.title.set_text(" x6[n] Reel")
imag.stem(n2, signal6.imag)
imag.title.set_text(" x6[n] Imag")
plt.tight_layout(pad=2.0)

```

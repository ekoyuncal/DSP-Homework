## Cevap A. Bu sistem kararlÄ± deÄŸildir Ã§Ã¼nkÃ¼   y [n] deÄŸeri((1.2)^n * u[n]) n arttÄ±kÃ§a sonsuze yaklaÅŸÄ±r |y[n]| -> âˆž


### b-)Bu sisteme x[n] = u[n-3] iÅŸaretini uygulayÄ±nÄ±z.

```python
n = np.arange(0,20)
x = np.heaviside(n-3, 1) - np.heaviside(n-7, 1)


plt.figure()
plt.stem(n, x)
plt.title("x[n] = u[n âˆ’ 3] iÅŸareti UygulanmÄ±ÅŸ Sinyal ")
plt.xlabel("Sample Number n")
plt.ylabel("Amplitude")

#ak ve bk katsayÄ±larÄ±nÄ±n tanÄ±mlanmasÄ±
a = np.array([1, -1.2], dtype= float)
b = np.array([0, 0, 0, 1, 1, 1, 1, 0, 0], dtype= float)#x[n] = u[n-3]

#lfilter iÅŸlemi
y = signal.lfilter(b, a , x)


#lflilter sinyalinin Ã§izdirilmesi
plt.figure()
plt.stem(n, y)
plt.title("Lfilter ile OluÅŸturulan y[n] Sinyal")
plt.xlabel("Sample Number n")
plt.ylabel("Amplitude")

#convolve iÅŸlemi
y2 = signal.convolve(y, x, "same")

#convolve edilmiÅŸ sinyalin Ã§izdirilmesi
plt.figure()
plt.stem(n, y2)
plt.title("Convolve ile OluÅŸturulan Sinyal")
plt.xlabel("Sample Number n")
plt.ylabel("Amplitude")
```

## Soru-2 a, b, c ÅŸÄ±klarÄ±nda verilen giriÅŸ iÅŸaretlerinin sisteme uygulanmasÄ± sonucu oluÅŸan giriÅŸ ve  Ã§Ä±kÄ±ÅŸ sinyallerini Ã§izdiriniz.

### Sistem-2 h2[n] = d[n-3] -d[n-4]

### a-) x3 GiriÅŸ ve Ã‡Ä±kÄ±ÅŸ Sinyali

### b-) x4 GiriÅŸ ve Ã‡Ä±kÄ±ÅŸ Sinyali

### c-) x5 GiriÅŸ ve Ã‡Ä±kÄ±ÅŸ Sinyali


```python
n = np.arange(-1, 25)
n2 = np.arange(-1, 11)
n3 = np.arange(-1, 25)

#sistem dÃ¼rtÃ¼ diraclarÄ±nÄ±n bulunduÄŸu noktalar ve deÄŸerleri
h1 = np.array([0, 0, 0, 1, -1, 0, 0], dtype = float)


#GiriÅŸ Sinyallerinin tanÄ±mlnamasÄ±
x3 = np.cos(np.pi/5 * n) * np.heaviside(n, 1) - np.cos(np.pi/5 * n) * np.heaviside(n-20, 1)

x4 = ((-0.3)**n2) *( np.heaviside(n2-1, 1) - np.heaviside(n2-5, 1))

x5 = n3*np.heaviside(n3, 1)

#Ã‡Ä±kÄ±ÅŸ Sinyallerinin tanÄ±mlanmasÄ±
y3 = np.convolve(x3, h1, "same")

y4 = np.convolve(x4, h1, "same")
 
#same parametresini kullanmaya Ã§alÄ±ÅŸtÄ±ÄŸÄ±mda
#elle hesapladÄ±ÄŸÄ±m deÄŸerler ve indisler uyuÅŸmuyor
y5 = np.convolve(x5, h1)


#GiriÅŸ Sinyallerinin Ã‡izdirilmesi
plt.figure()
plt.stem(n, x3)
plt.title("x3[n] GiriÅŸ Sinyali")
plt.xlabel("Sample Number n")
plt.ylabel("Amplitude")

plt.figure()
plt.stem(n2, x4)
plt.title("x4[n] GiriÅŸ Sinyali")
plt.xlabel("Sample Number n")
plt.ylabel("Amplitude")

plt.figure()
plt.stem(n3, x5)
plt.title("x5[n] GiriÅŸ Sinyali")
plt.xlabel("Sample Number n")
plt.ylabel("Amplitude")




#Ã‡Ä±kÄ±ÅŸ Sinyallerinin Ã‡izdirilmesi
plt.figure()
plt.stem(n, y3)
plt.title("y3[n] Ã‡Ä±kÄ±ÅŸ Sinyali")
plt.xlabel("Sample Number n")
plt.ylabel("Amplitude")

plt.figure()
plt.stem(n2, y4)
plt.title("y4[n] Ã‡Ä±kÄ±ÅŸ Sinyali")
plt.xlabel("Sample Number n")
plt.ylabel("Amplitude")

#same kullanmadÄ±ÄŸÄ±m iÃ§in n3 sample uzunluÄŸu y5 uzunluÄŸu ile uyuÅŸmadÄ±
#bu yÃ¼zden y5 i kendi indisleri ile Ã§izdiriyorum.
plt.figure()
plt.stem(y5)
plt.title("y5[n] Ã‡Ä±kÄ±ÅŸ Sinyali")
plt.xlabel("Sample Number n")
plt.ylabel("Amplitude")

```

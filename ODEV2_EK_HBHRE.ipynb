```python
#Kütüphanelerin import edilmesi
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
%matplotlib notebook
```

## Soru-1 𝑦[𝑛] − 1.2𝑦[𝑛 − 1] = 𝑥[𝑛] − 𝑥[𝑛 − 4] Sistemi için aşağıda verilen soruları yanıtlayınız.

### a-)Bu sistem kararlı mıdır? neden ?

## Cevap A. Bu sistem kararlı değildir çünkü   y [n] değeri((1.2)^n * u[n]) n arttıkça sonsuze yaklaşır |y[n]| -> ∞


### b-)Bu sisteme x[n] = u[n-3] işaretini uygulayınız.

```python
n = np.arange(0,20)
x = np.heaviside(n-3, 1) - np.heaviside(n-7, 1)


plt.figure()
plt.stem(n, x)
plt.title("x[n] = u[n − 3] işareti Uygulanmış Sinyal ")
plt.xlabel("Sample Number n")
plt.ylabel("Amplitude")

#ak ve bk katsayılarının tanımlanması
a = np.array([1, -1.2], dtype= float)
b = np.array([0, 0, 0, 1, 1, 1, 1, 0, 0], dtype= float)#x[n] = u[n-3]

#lfilter işlemi
y = signal.lfilter(b, a , x)


#lflilter sinyalinin çizdirilmesi
plt.figure()
plt.stem(n, y)
plt.title("Lfilter ile Oluşturulan y[n] Sinyal")
plt.xlabel("Sample Number n")
plt.ylabel("Amplitude")

#convolve işlemi
y2 = signal.convolve(y, x, "same")

#convolve edilmiş sinyalin çizdirilmesi
plt.figure()
plt.stem(n, y2)
plt.title("Convolve ile Oluşturulan Sinyal")
plt.xlabel("Sample Number n")
plt.ylabel("Amplitude")
```

## Soru-2 a, b, c şıklarında verilen giriş işaretlerinin sisteme uygulanması sonucu oluşan giriş ve  çıkış sinyallerini çizdiriniz.

### Sistem-2 h2[n] = d[n-3] -d[n-4]

### a-) x3 Giriş ve Çıkış Sinyali

### b-) x4 Giriş ve Çıkış Sinyali

### c-) x5 Giriş ve Çıkış Sinyali


```python
n = np.arange(-1, 25)
n2 = np.arange(-1, 11)
n3 = np.arange(-1, 25)

#sistem dürtü diraclarının bulunduğu noktalar ve değerleri
h1 = np.array([0, 0, 0, 1, -1, 0, 0], dtype = float)


#Giriş Sinyallerinin tanımlnaması
x3 = np.cos(np.pi/5 * n) * np.heaviside(n, 1) - np.cos(np.pi/5 * n) * np.heaviside(n-20, 1)

x4 = ((-0.3)**n2) *( np.heaviside(n2-1, 1) - np.heaviside(n2-5, 1))

x5 = n3*np.heaviside(n3, 1)

#Çıkış Sinyallerinin tanımlanması
y3 = np.convolve(x3, h1, "same")

y4 = np.convolve(x4, h1, "same")
 
#same parametresini kullanmaya çalıştığımda
#elle hesapladığım değerler ve indisler uyuşmuyor
y5 = np.convolve(x5, h1)


#Giriş Sinyallerinin Çizdirilmesi
plt.figure()
plt.stem(n, x3)
plt.title("x3[n] Giriş Sinyali")
plt.xlabel("Sample Number n")
plt.ylabel("Amplitude")

plt.figure()
plt.stem(n2, x4)
plt.title("x4[n] Giriş Sinyali")
plt.xlabel("Sample Number n")
plt.ylabel("Amplitude")

plt.figure()
plt.stem(n3, x5)
plt.title("x5[n] Giriş Sinyali")
plt.xlabel("Sample Number n")
plt.ylabel("Amplitude")




#Çıkış Sinyallerinin Çizdirilmesi
plt.figure()
plt.stem(n, y3)
plt.title("y3[n] Çıkış Sinyali")
plt.xlabel("Sample Number n")
plt.ylabel("Amplitude")

plt.figure()
plt.stem(n2, y4)
plt.title("y4[n] Çıkış Sinyali")
plt.xlabel("Sample Number n")
plt.ylabel("Amplitude")

#same kullanmadığım için n3 sample uzunluğu y5 uzunluğu ile uyuşmadı
#bu yüzden y5 i kendi indisleri ile çizdiriyorum.
plt.figure()
plt.stem(y5)
plt.title("y5[n] Çıkış Sinyali")
plt.xlabel("Sample Number n")
plt.ylabel("Amplitude")

```
